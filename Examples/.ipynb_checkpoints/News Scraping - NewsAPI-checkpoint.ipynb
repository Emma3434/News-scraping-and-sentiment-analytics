```python
import requests
import pandas as pd
import newsapi
```

```python
query = "Apple"
ticker = "AAPL"
from_date = "2021-03-01"
to_date = "2021-03-13"
sortBy = "relevancy"
apiKey = "4dbc17e007ab436fb66416009dfb59a8"
```

```python
news_api_url = ('https://newsapi.org/v2/everything?'
       'q='+query+
       '&from='+from_date+
       '&to='+to_date+
       '&sortBy='+sortBy+
       '&apiKey='+apiKey+
               '&page=2')
res = requests.get(news_api_url)

page= res.json()
df = pd.DataFrame(page['articles'])
```
